```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[23,5,6,79,10]

plt.plot(x, y, marker = '*')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Simple Line Plot")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(11)
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
temp = np.random.randint(20,50,7)

plt.plot(days, temp, marker = "*")

plt.xlabel("Days ----->")
plt.ylabel("Temp ----->")
plt.title("Variation in temperature for one week")
plt.xticks(rotation=45)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(11)
days = list(range(1,31))
temp = np.random.randint(20,50,30)

plt.figure(figsize=(8,6))

plt.bar(days, temp, color = ["b","g","y","r"],edgecolor="black", alpha=0.7)

plt.xlabel("Days ----->")
plt.ylabel("Temp ----->")
plt.title("Variation in temperature in a month")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
