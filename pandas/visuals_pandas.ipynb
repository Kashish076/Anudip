```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "Months":['jan','feb','march','april','may'],
    "Sales":[5500,2334,4500,3500,8000]
}
df = pd.DataFrame(data)
df.set_index("Months",inplace=True)
df.plot(kind="line")
plt.show()
```
